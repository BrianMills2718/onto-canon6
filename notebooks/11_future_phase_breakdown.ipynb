# Future Phase Breakdown

This notebook fills in the planned phases after Phase 9.

Phase 9 is already proved in `notebooks/15_epistemic_extension_slice.ipynb`.
This notebook now focuses on Phase 10.

It is meant to be a practical planning companion to `docs/plans/0001_successor_roadmap.md`:

1. the remaining future phase has a goal, build order, non-goals, acceptance criteria, and explicit uncertainties;
2. the notebook includes lightweight sketches so the future direction is easier to inspect;
3. the roadmap remains the authoritative prose record.


```python
from __future__ import annotations

from dataclasses import dataclass
from pprint import pprint


@dataclass(frozen=True)
class FuturePhase:
    phase_id: str
    title: str
    goal: str
    build_order: tuple[str, ...]
    non_goals: tuple[str, ...]
    acceptance: tuple[str, ...]
    uncertainties: tuple[str, ...]


def show_phase(phase: FuturePhase) -> None:
    print(f"\nPhase {phase.phase_id}: {phase.title}")
    print("=" * (len(phase.title) + len(phase.phase_id) + 8))
    print(f"goal: {phase.goal}\n")
    print("build order:")
    for item in phase.build_order:
        print(f"- {item}")
    print("\nnon-goals:")
    for item in phase.non_goals:
        print(f"- {item}")
    print("\nacceptance criteria:")
    for item in phase.acceptance:
        print(f"- {item}")
    print("\nuncertainties:")
    for item in phase.uncertainties:
        print(f"- {item}")
```

```python
phases = {
    "10": FuturePhase(
        phase_id="10",
        title="Product-Facing Workflow Integration",
        goal="Connect the proved stack into one real end-to-end workflow with user-visible leverage.",
        build_order=(
            "Choose one workflow and one integration boundary explicitly before building anything around it.",
            "Reuse the proved CLI/review/extraction/overlay/artifact/epistemic services rather than creating a new workflow runtime.",
            "Add only the minimal outward-facing adapter needed for that workflow.",
            "Prove the workflow with real inputs and explicit provenance traces.",
        ),
        non_goals=(
            "Multiple product workflows at once.",
            "A generalized platform or multi-tenant service.",
            "New architecture layers that exist only for hypothetical later products.",
        ),
        acceptance=(
            "A user can complete one real workflow end to end without ad hoc Python editing.",
            "The workflow shows clear leverage over the pre-successor state.",
            "The workflow does not reintroduce a fused runtime blob.",
        ),
        uncertainties=(
            "Which single workflow best demonstrates real user-visible leverage?",
            "Should the first outward-facing boundary after CLI be MCP or something narrower?",
            "What minimum provenance depth is required before the workflow feels credibly useful?",
        ),
    ),
}

for phase_id in ("10",):
    show_phase(phases[phase_id])
```
