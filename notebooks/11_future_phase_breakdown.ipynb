# Future Phase Breakdown

This notebook fills in the planned phases after Phase 5.

It is meant to be a practical planning companion to `docs/plans/0001_successor_roadmap.md`:

1. each phase has a goal, build order, non-goals, acceptance criteria, and explicit uncertainties;
2. the notebook includes a few lightweight sketches so the future direction is easier to inspect;
3. the roadmap remains the authoritative prose record.


```python
from __future__ import annotations

from dataclasses import dataclass
from pprint import pprint


@dataclass(frozen=True)
class FuturePhase:
    phase_id: str
    title: str
    goal: str
    build_order: tuple[str, ...]
    non_goals: tuple[str, ...]
    acceptance: tuple[str, ...]
    uncertainties: tuple[str, ...]


def show_phase(phase: FuturePhase) -> None:
    print(f"\nPhase {phase.phase_id}: {phase.title}")
    print("=" * (len(phase.title) + len(phase.phase_id) + 8))
    print(f"goal: {phase.goal}\n")
    print("build order:")
    for item in phase.build_order:
        print(f"- {item}")
    print("\nnon-goals:")
    for item in phase.non_goals:
        print(f"- {item}")
    print("\nacceptance criteria:")
    for item in phase.acceptance:
        print(f"- {item}")
    print("\nuncertainties:")
    for item in phase.uncertainties:
        print(f"- {item}")
```

```python
phases = {
    "6": FuturePhase(
        phase_id="6",
        title="First Operational Surface",
        goal="Make the proved workflow usable without direct Python or notebook editing.",
        build_order=(
            "Define the CLI command surface and argument contract before wiring handlers.",
            "Implement thin handlers for extract, list, review, and apply actions.",
            "Add JSON output first, then human-readable output on top of the same data.",
            "Add integration tests and one transcript/notebook proof last.",
        ),
        non_goals=(
            "MCP or UI surfaces.",
            "Background jobs or remote deployment.",
            "Direct database manipulation from CLI handlers.",
        ),
        acceptance=(
            "A user can go from raw text to persisted candidate assertions through the CLI.",
            "A user can inspect candidates and proposals, record review decisions, and apply overlays through the CLI.",
            "CLI handlers delegate to existing services rather than owning workflow logic.",
        ),
        uncertainties=(
            "Should the first extract command accept files only, or also stdin/raw literals?",
            "Should human-readable output be table-first or JSON-first?",
            "Should candidate and proposal review stay as separate command groups?",
        ),
    ),
    "7": FuturePhase(
        phase_id="7",
        title="Domain Pack Generalization",
        goal="Prove that the same architecture supports a second real domain pack.",
        build_order=(
            "Choose the minimal second-pack scope before importing or rewriting ontology material.",
            "Define one strict profile and one mixed profile for that pack.",
            "Prove validation, proposal routing, and overlay behavior locally.",
            "Add pack-specific extraction context only if the current prompts are insufficient.",
        ),
        non_goals=(
            "Full DoDAF coverage.",
            "Cross-pack inference or ontology merger as a new subsystem.",
            "Domain-specific branches in core modules.",
        ),
        acceptance=(
            "The second pack loads through the same pack/profile machinery as existing packs.",
            "The same review and overlay workflow works for the second pack.",
            "No domain-specific logic is added to core or ontology runtime modules.",
        ),
        uncertainties=(
            "Which exact DoDAF subset is the right first proof surface?",
            "Should the second pack be imported from donor material or authored locally from a reduced subset?",
            "Is a shared base vocabulary layer needed before the second pack stays clean?",
        ),
    ),
    "8": FuturePhase(
        phase_id="8",
        title="Artifact Lineage Recovery",
        goal="Recover artifact-backed provenance from onto-canon v1 without rebuilding a fused runtime.",
        build_order=(
            "Define the minimal typed artifact model and persistence boundary.",
            "Link artifacts to candidate assertions and review records first.",
            "Add derived-artifact support and lineage reporting second.",
            "Prove one workflow where a claim is supported by an analysis artifact rather than only raw text.",
        ),
        non_goals=(
            "A full artifact warehouse or blob store.",
            "Workflow scheduling or orchestration.",
            "A generalized research-platform artifact system.",
        ),
        acceptance=(
            "Assertions can reference both raw sources and derived analysis artifacts.",
            "Lineage queries do not require moving artifact logic into core.",
            "Provenance remains explicit and inspectable rather than hidden in metadata blobs.",
        ),
        uncertainties=(
            "Which artifact categories are essential for the first slice?",
            "Should artifact links terminate at candidate assertions only or also future accepted-assertion projections?",
            "How much artifact deduplication is worth doing before real pressure exists?",
        ),
    ),
    "9": FuturePhase(
        phase_id="9",
        title="Epistemic Extension",
        goal="Add optional epistemic reasoning without collapsing subsystem boundaries.",
        build_order=(
            "Choose the smallest useful epistemic operator set before defining storage.",
            "Implement extension-local models and operators.",
            "Wire them through explicit seams from existing review/provenance records.",
            "Prove that the base workflow still functions unchanged when the extension is disabled.",
        ),
        non_goals=(
            "A full truth-maintenance system.",
            "Automatic contradiction resolution across the whole graph.",
            "Recentering the runtime around an epistemic controller.",
        ),
        acceptance=(
            "Epistemic data stays optional and does not become a hidden requirement for the base workflow.",
            "The extension depends only on explicit seams and existing persisted records.",
            "The extension has its own tests and notebook proof.",
        ),
        uncertainties=(
            "Should the first extension slice start with confidence, tension, supersession, or contradiction?",
            "Should epistemic state attach only to accepted assertions or also candidate assertions under review?",
            "Are confidence values user-entered, model-derived, or both?",
        ),
    ),
    "10": FuturePhase(
        phase_id="10",
        title="Product-Facing Workflow Integration",
        goal="Connect the proved stack into one real end-to-end workflow with user-visible leverage.",
        build_order=(
            "Choose one workflow and one integration boundary explicitly before building anything around it.",
            "Reuse the proved CLI/review/extraction/overlay/artifact services rather than creating a new workflow runtime.",
            "Add only the minimal outward-facing adapter needed for that workflow.",
            "Prove the workflow with real inputs and explicit provenance traces.",
        ),
        non_goals=(
            "Multiple product workflows at once.",
            "A generalized platform or multi-tenant service.",
            "New architecture layers that exist only for hypothetical later products.",
        ),
        acceptance=(
            "A user can complete one real workflow end to end without ad hoc Python editing.",
            "The workflow shows clear leverage over the pre-successor state.",
            "The workflow does not reintroduce a fused runtime blob.",
        ),
        uncertainties=(
            "Which single workflow best demonstrates real user-visible leverage?",
            "Should the first outward-facing boundary after CLI be MCP or something narrower?",
            "What minimum provenance depth is required before the workflow feels credibly useful?",
        ),
    ),
}

for phase_id in ("6", "7", "8", "9", "10"):
    show_phase(phases[phase_id])
```
